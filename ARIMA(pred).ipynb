```python
#importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as dates
from datetime import datetime

# Extra Libs
import matplotlib.dates as mdates
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
from bokeh.models import HoverTool
from IPython.display import HTML, display

import statsmodels.api as sm
```

```python
df=pd.read_csv('/content/sample_data/230922_MK_heartrate.csv')
df.head()
```

```python
df = df[['update_time', 'heart_rate']]
df.head(10)
```

```python
# update_time -> 날짜변수로 변경
df = df.copy()
df['update_time'] = pd.to_datetime(df['update_time'])
```

```python
#날짜순 정렬(ascending=T : 오름차순)
df = df.sort_values(by='update_time', ascending=True)

#idx,date, month, year 변수 생성
dataset_length = len(df)
df['idx'] = range(1, dataset_length + 1)
df['date'] = df['update_time'].dt.date
df['month'] = df['update_time'].dt.month
df['year'] = df['update_time'].dt.year

df.head()
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,4)
plt.plot(df['date'],df['heart_rate'])
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y.%m'))
plt.show()
```

```python
#train:test=9:1 나누기
split_ratio = 0.9
split_index = int(len(df) * split_ratio)

df_train = df[:split_index]
df_test = df[split_index:]


df_train.head(5)
```

# Test #

##ADF test##

```python
from statsmodels.tsa.stattools import adfuller
#Ho: 정상성 만족X
#H1: 정상성 만족O
# A function defined for Dickey Fuller Test
def adfuller_test(births):
    result=adfuller(births)     #adf검정
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:    #p : result[1] <= 0.05
        print("정상성 만족O")
    else:
        print("정상성 만족X ")
```

```python
adfuller_test(df_train['heart_rate'])
```

#ACF, PACF#

## diff(0)##

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.tools import diff

plt.figure(figsize=(16,4))
ax1 = plt.subplot(1,2,1)
plot_acf(df_train['heart_rate'], title='ACF',ax=ax1 )
plt.xlabel('Lag')
plt.ylabel('ACF')
ax2 = plt.subplot(1,2,2)
plot_pacf(df_train['heart_rate'], title='PACF',ax=ax2)
plt.xlabel('Lag')
plt.ylabel('Partial ACF')
plt.show()
```

ACF: lag=34를 기준으로 상관관계가 양에서 음으로 변동     -> p = 1

PACF : 1ag=1에서 pacf=0.65정도   -> q=2

##diff(1)##

```python
plt.figure(figsize=(16,4))
ax1 = plt.subplot(1,2,1)
plot_acf(diff(df_train['heart_rate']), title='ACF',ax=ax1 ,use_vlines=True)
plt.xlabel('Lag')
plt.ylabel('ACF')
ax2 = plt.subplot(1,2,2)
plot_pacf(diff(df_train['heart_rate']), title='PACF',ax=ax2)
plt.xlabel('Lag')
plt.ylabel('Partial ACF')
plt.show()
```
